## Feature Engineering

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"]
)
```

```python
df["Month"] = df["Order Date"].dt.month

df["Year"] = df["Order Date"].dt.year
```

```python
monthly_sales = df.groupby(
    "Month"
)["Sales"].sum()

monthly_sales.plot(
    kind="line",
    figsize=(10,5),
    marker="o"
)

plt.title("Monthly Sales Trend")

plt.ylabel("Sales")

plt.show()
```

```python
df = df[
    (df["Sales"] < 2000) &
    (df["Profit"] > -1000) &
    (df["Profit"] < 1000)
].copy()

print(df.shape)
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Profit"])

plt.title("Profit After Outlier Treatment")

plt.show()
```

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
from sklearn.model_selection import (train_test_split, cross_val_score)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
categorical_cols = ["Segment", "Region", "Category", "Sub-Category"]
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
columns_to_use = ['Sales', 'Quantity', 'Discount', 'Month', 'Year', 'Segment', 'Region', 'Category', 'Profit']
df_simple = df[columns_to_use]

df_encoded = pd.get_dummies(df_simple, drop_first=True)
```

```python
X = df_encoded.drop(columns=['Profit'])
y = df_encoded['Profit']

X_train = X[X['Year'] <= 2016]
y_train = y[X['Year'] <= 2016]

X_test = X[X['Year'] == 2017]
y_test = y[X['Year'] == 2017]
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

y_pred_dt = dt_model.predict(X_test)
```

```python
rf_r2 = r2_score(y_test, y_pred_rf)
lr_r2 = r2_score(y_test, y_pred_lr)
dt_r2 = r2_score(y_test, y_pred_dt)
```

```python
rf_mae = mean_absolute_error(y_test, y_pred_rf)
lr_mae = mean_absolute_error(y_test, y_pred_lr)
dt_mae = mean_absolute_error(y_test, y_pred_dt)
```

```python
rf_rmse = np.sqrt(mean_squared_error(y_test,y_pred_rf))
lr_rmse = np.sqrt(mean_squared_error(y_test,y_pred_lr))
dt_rmse = np.sqrt(mean_squared_error(y_test,y_pred_dt))
```

```python
print(f"Random Forest R² Score: {rf_r2:.2f}")
print(f"Linear Regression R² Score: {lr_r2:.2f}")
print(f"Decision Tree R² Score: {dt_r2:.2f}")
```

```python
mae = mean_absolute_error(y_test, y_pred_rf)

print("MAE:", round(mae,2))
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred_rf))

print("RMSE:", round(rmse,2))
```

```python
print(f"Linear Regression R² Score: {lr_r2 * 100:.2f}%")
print(f"Decision Tree R² Score: {dt_r2 * 100:.2f}%")
print(f"Random Forest R² Score: {rf_r2 * 100:.2f}%\n")

print(f"Linear Regression MAE: {lr_mae:.2f}")
print(f"Decision Tree MAE: {dt_mae:.2f}")
print(f"Random Forest MAE: {rf_mae:.2f}\n")

print(f"Linear Regression RMSE: {lr_rmse:.2f}")
print(f"Decision Tree RMSE: {dt_rmse:.2f}")
print(f"Random Forest RMSE: {rf_rmse:.2f}")
```

```python
print(f"Linear Regression Train R² Score: {lr_model.score(X_train, y_train)*100:.2f}%")
print(f"Linear Regression Test R² Score: {lr_model.score(X_test, y_test)*100:.2f}%\n")

print(f"Decision Tree Train R² Score: {dt_model.score(X_train, y_train)*100:.2f}%")
print(f"Decision Tree Test R² Score: {dt_model.score(X_test, y_test)*100:.2f}%\n")

print(f"Random Forest Train R² Score: {rf_model.score(X_train, y_train)*100:.2f}%")
print(f"Random Forest Test R² Score: {rf_model.score(X_test, y_test)*100:.2f}%")
```

```python
print("Cross Validation Results")
print("-"*90)

lr_cv = cross_val_score(lr_model, X, y, cv=5, scoring="r2")
print("\nLinear Regression CV Scores:", lr_cv)
print(f"Linear Regression Average CV R²: {lr_cv.mean():.4f}\n")

dt_cv = cross_val_score(dt_model, X, y, cv=5, scoring="r2")
print("Decision Tree CV Scores:", dt_cv)
print(f"Decision Tree Average CV R²: {dt_cv.mean():.4f}\n")

rf_cv = cross_val_score(rf_model, X, y, cv=5, scoring="r2")
print("Random Forest CV Scores:", rf_cv)
print(f"Random Forest Average CV R²: {rf_cv.mean():.4f}")
```

```python
results = pd.DataFrame(
{
    "Model":["Linear Regression", "Decision Tree", "Random Forest"],
    "R2 Score":[lr_r2, dt_r2, rf_r2]
}
)

results.sort_values(
    by="R2 Score",
    ascending=False
)
```
